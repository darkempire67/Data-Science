```python
# A --------------Reading the iris data set from a url and assigning it to DataFrame


# The following line will import KNeighborsClassifier "Class"
# KNeighborsClassifier is name of a "sklearn class" to perform "KNN Classification" 

from sklearn.neighbors import KNeighborsClassifier
# Importing the required packages and libraries
# we will need numpy and pandas later
import numpy as np
import pandas as pd

# reading dataset from link
iris_df = pd.read_csv('https://raw.githubusercontent.com/mpourhoma/CS4661/master/iris.csv')

# features that we will use for this part
feature_cols = ['sepal_length','sepal_width','petal_length','petal_width']
X= iris_df[feature_cols]  # defining the features for our data X

# select a Series of labels (the last column) from the DataFrame
# y = iris_df['label'] # this is the index that we gave to the labels
# OR:
y = iris_df['species'] # this is the original categorical labels (the latest revision of sklearn accepts non-numerical labels)
#--print(iris_df)

# A ---------------------------------
```

```python
# B ------- Splitting the dataset into testing and traing sets using tes_size=.4 and random_state= 6
#----------------- 

# Randomly splitting the original dataset into training set and testing set
# The function"train_test_split" from "sklearn.cross_validation" library performs random splitting.
# "test_size=0.4" means that pick 40% of data samples for testing set, and the rest (60%) for training set.


from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=6)
print (X_test.shape)
print (y_test.shape)
# B --------------------------
```

```python
# C ---------- Instatiating KNN object with k=3 testing, training and calculating accuracy.


# Instantiating another "object" of KNeighborsClassifier "class" with k=3:

k = 3
my_knn = KNeighborsClassifier(n_neighbors=k) # name of the object is arbitrary!
my_knn.fit(X_train, y_train)
y_predict =my_knn.predict(X_test)
#print(y_predict)
# We can now compare the "predicted labels" for the Testing Set with its "actual labels" to evaluate the accuracy 
# Function "accuracy_score" from "sklearn.metrics" will perform the element-to-element comparision and returns the 
# portion of correct predictions:

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_predict)  # testing the accuracy
results=pd.DataFrame()                        # saving prediction and results to DataFrame 
results['actual']=y_test                      # saving accuracy to variable
results['prediction']=y_predict
print(results,'\n',accuracy)


#   -------------------------------------------
# C -----  the accuracy is 0.9666666666666667
```

```python
# D ------- Repeating part C but with different K values: (1,5,7,11,15,27,59)


list_of_k=[1,5,7,11,15,27,59] # the k we need to calculate
all_accuracy_results={}       # we will save accuracy results in dict
from sklearn.metrics import accuracy_score

all_results=pd.DataFrame()   # results comprison will be saved in a DF()
for k in range(1,60):        # loop through the range 
    if k in list_of_k:       # looking for our k's
        
        k_ = k 
        # In the following line, "knn" is instantiated as an "object" of KNeighborsClassifier "class". 
        my_knn = KNeighborsClassifier(n_neighbors=k_) # name of the object is arbitrary!
            
        # We can use the method "fit" of the "object knn" along with training dataset and labels to train the model.
        my_knn.fit(X_train, y_train)      # Training ONLY on the training set:
            
        # We can use the method "predict" of the *trained* object knn on one or more testing data sample to perform prediction:
        y_predict =my_knn.predict(X_test) # Testing on the testing set:
        #--print(y_predict)
        # We can now compare the "predicted labels" for the Testing Set with its "actual labels" to evaluate the accuracy 
        # Function "accuracy_score" from "sklearn.metrics" will perform the element-to-element comparision and returns the 
        # portion of correct predictions:
        accuracy = accuracy_score(y_test, y_predict)
        
        
        all_results['actual ',k_]=y_test           # saving results and prediction to DataFrame
        all_results['prediction ',k_]=y_predict
        
        #--print(results)
        kk='k= '+str(k) # string to create the name of that k value 'k = 1' or 'k = 2'
        
        all_accuracy_results[kk]=accuracy # Dict to add "k num " and Accuracy number into dict{k,v} as value
        
        
print(all_accuracy_results) # Accuracy gets worst the higher K is
print (all_results) # dataframe that contains results

# -----------------------------------------------------------------------------------------
# D ---- The Accuracy gets WORSE if K is increased too much.
```

```python
# E ----- Based on One feature ,k=3, using each feature- we will build four models &-
# --------we will see what is the best feature out of the four features based on the accuracy.

feature_cols_list = ['sepal_length','sepal_width','petal_length','petal_width']
accuracy_results_oneF={}    # saving accuracy data to list
results_oneF=pd.DataFrame() # saving results to DataFrame obj
from sklearn.model_selection import train_test_split


for f in feature_cols_list:   # loop through list and find resusts and accuracy for each element
    print(f)
    X= iris_df[[f]]   # using one feature at a time using [] for X, else Error.
    #--print(iris_df)
    #--print(X)
    
    #splitting the data using method from sklearn.
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=6)
    #--print (X_test)
    #--print (y_test)
    # In the following line, "knn" is instantiated as an "object" of KNeighborsClassifier "class". 
    k = 3
    my_knn = KNeighborsClassifier(n_neighbors=k) # name of the object is arbitrary!
    
    # We can use the method "fit" of the "object knn" along with training dataset and labels to train the model.
    my_knn.fit(X_train, y_train)
    
    # We can use the method "predict" of the *trained* object knn on one or more testing data sample to perform prediction:
    y_predict =my_knn.predict(X_test)
    #--print(y_predict)
    # We can now compare the "predicted labels" for the Testing Set with its "actual labels" to evaluate the accuracy 
    # Function "accuracy_score" from "sklearn.metrics" will perform the element-to-element comparision and returns the 
    # portion of correct predictions:

    from sklearn.metrics import accuracy_score

    accuracy = accuracy_score(y_test, y_predict)  # saving Accuracy results to list
    accuracy_results_oneF[f]=accuracy
    
    results_oneF['actual for ',f]=y_test      # saving the prediction and results in DataFrame
    results_oneF['prediction for',f]=y_predict
print(results_oneF,'\n\n',accuracy_results_oneF)


# -----------------------------------------------------------------------------------------
# E ------- Best feature to provide the Best Accuracy is 'petal_width': 0.95
# E ------- Second best is 'petal_length': 0.93
```

```python
# F----- Repeating part E but instead of using One feature we wil use Two features.
#--- A Permutation of features list if you will.


#feature_cols = ['sepal_length','sepal_width','petal_length','petal_width']  orignal feat_col

p_feat_cols=[['sepal_length','sepal_width'],['sepal_length','petal_length'],['sepal_length','petal_width'],['sepal_width','petal_length'],['sepal_width','petal_width'],['petal_length','petal_width']]

#print(p_feat_cols[0])
accuracy_results_p={}         # Accuracy results will be saved in list.
results_perm=pd.DataFrame()   # Results results will be saved in DataFrame.
from sklearn.model_selection import train_test_split

for twoF in p_feat_cols:   # loop throught the permuted list to get two features at once.
    #--print(twoF)
    X= iris_df[twoF]    # using two features at a time
    
    #--print(X)
    
    y = iris_df['species'] # this is the original categorical labels (the latest revision of sklearn accepts non-numerical labels)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=6)
    #--print (X_test)
    #--print (y_test)
    
    # In the following line, "knn" is instantiated as an "object" of KNeighborsClassifier "class". 
    k = 3
    my_knn = KNeighborsClassifier(n_neighbors=k) # name of the object is arbitrary!
    
    # We can use the method "fit" of the "object knn" along with training dataset and labels to train the model.
    my_knn.fit(X_train, y_train)
    # We can use the method "predict" of the *trained* object knn on one or more testing data sample to perform prediction:
    y_predict =my_knn.predict(X_test)
    
    #--print(y_predict)
    
    # We can now compare the "predicted labels" for the Testing Set with its "actual labels" to evaluate the accuracy 
    # Function "accuracy_score" from "sklearn.metrics" will perform the element-to-element comparision and returns the 
    # portion of correct predictions:

    from sklearn.metrics import accuracy_score
     
    accuracy = accuracy_score(y_test, y_predict)             #    accuracy = accuracy_score(y_test, y_predict)             
    accuracy_results_p[" and ".join(twoF)]=accuracy        # saving Accuracy rusults to a list
    
    
    results_perm['actual for '," and ".join(twoF)]=y_test     # adding results and predictions to DataFrame obj
    results_perm['prediction for'," and ".join(twoF)]=y_predict
print(results_perm,'\n\n',accuracy_results_p)
# -----------------------------------------------------------------------------------------
# F -------BEST Accruracy with TWo Features is with ...sepal_length and petal_length': 0.9833
```
